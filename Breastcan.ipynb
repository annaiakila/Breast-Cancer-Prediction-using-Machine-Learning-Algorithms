Over Sampling

```python
from sklearn.utils import resample

df_majority=df[df.Breast == 0]
df_minority=df[df.Breast == 1]

df_minority_oversampled = resample(df_minority,
                                   replace=True,
                                   n_samples=len(df_majority),
                                   random_state=42)

df_oversampled=pd.concat([df_majority, df_minority_oversampled])

df_oversampled['Breast'].value_counts(normalize=True)
```

Split into X and Y

```python
# Cleanup + prepare features and target
# Drop 'S/N' (serial number) as it's not a predictive feature
df = df.copy()
if 'S/N' in df.columns:
    df = df.drop(columns=['S/N'])
# Convert numeric-like columns to numeric (remove non-digits like '#', '<', '>')
numeric_like = ['Year', 'Age', 'Tumor Size (cm)', 'Inv-Nodes']
for col in numeric_like:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace('[^0-9.-]', '', regex=True), errors='coerce')
# Identify categorical columns that need encoding
categorical_cols = ['Menopause', 'Metastasis', 'History', 'Breast', 'Breast Quadrant']
from sklearn.preprocessing import LabelEncoder
for col in categorical_cols:
    if col in df.columns:
        if df[col].dtype == object or df[col].dtype.name == 'category':
            le = LabelEncoder()
            df[col] = le.fit_transform(df[col].astype(str))
# Impute missing numeric values with median
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
num_cols = df.select_dtypes(include=np.number).columns.tolist()
df[num_cols] = imputer.fit_transform(df[num_cols])
# Final check: list numeric columns and any NaNs
print('Numeric columns:', num_cols)
print('Any NaNs in numeric columns?', df[num_cols].isna().any().any())
# Split into X and Y
X = df.drop(columns=['Diagnosis Result'])
Y = df[['Diagnosis Result']]
X.shape, Y.shape
```

Train Test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
Y_test.shape
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

lr_model=LogisticRegression()
lr_model.fit(X_train,Y_train)

y_lr_pred=lr_model.predict(X_test)

print('accuracy score of logistic regression: ',accuracy_score(y_lr_pred,Y_test))
print('confusion matrix of logistic regression:')
print(confusion_matrix(y_lr_pred,Y_test))
print('classification report of logistic regression: ',classification_report(y_lr_pred,Y_test))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt_model=DecisionTreeClassifier()
dt_model.fit(X_train,Y_train)

y_dt_pred=dt_model.predict(X_test)

print('accuracy score of Decision Tree: ',accuracy_score(y_dt_pred,Y_test))
print('confusion matrix of Decision Tree: ')
print(confusion_matrix(y_lr_pred,Y_test))
print('classification report of Decision Tree: ',classification_report(y_lr_pred,Y_test))
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_model=RandomForestClassifier()
rf_model.fit(X_test,Y_test)

y_rf_pred=rf_model.predict(X_test)

print('accuracy score of Random Forest: ',accuracy_score(y_rf_pred,Y_test))
print('confusion matrix of Random Forest: ')
print(confusion_matrix(y_rf_pred,Y_test))
print('classification report of Random Forest: ',classification_report(y_rf_pred,Y_test))
```

Support Vector Machine

```python
from sklearn.svm import SVC

svm_model=SVC()
svm_model.fit(X_test,Y_test)

y_svm_pred=svm_model.predict(X_test)

print('accuracy score of Support vector machine: ',accuracy_score(y_svm_pred,Y_test))
print('confusion matrix of support vector machine: ')
print(confusion_matrix(y_svm_pred,Y_test))
print('classification report of Support vector machine: ',classification_report(y_svm_pred,Y_test))
```

Hyper Parameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier 

param_grid={
    'n_estimators':[100,200,300],
    'max_depth':[None,10,20,30],
    'min_samples_split':[2,5],
}

model=RandomForestClassifier()

grid=GridSearchCV(model,param_grid,cv=5,scoring='accuracy')
grid.fit(X_train,Y_train)

best_rf=grid.best_estimator_
y_r_pred=best_rf.predict(X_test)

print('Best Parameters: ',grid.best_params_)
print('Accuracy: ',accuracy_score(Y_test,y_r_pred))
print(classification_report(Y_test,y_r_pred))
print(confusion_matrix(Y_test,y_r_pred))
```
